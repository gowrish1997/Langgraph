```python
from langgraph.graph import StateGraph,START,END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.9)
```

```python
class BlogPostState(TypedDict):
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: BlogPostState) -> BlogPostState:
    outline = model.invoke(f"Create an outline for a blog post titled '{state['title']}'")
    state["outline"] = outline
    return state
def write_content(state: BlogPostState) -> BlogPostState:
    content = model.invoke(f"Write a blog post based on the following outline:\n{state['outline']}")
    state["content"] = content
    return state    
```

```python
graph = StateGraph(BlogPostState)

graph.add_node("create_outline",create_outline)
graph.add_node("write_content",write_content)

graph.add_edge(START,"create_outline")
graph.add_edge("create_outline","write_content")
graph.add_edge("write_content",END)
workflow = graph.compile()
```

```python
intial_state = {"title":"The Future of AI"}
final_state=workflow.invoke(intial_state)
print(final_state["content"].content)
```
